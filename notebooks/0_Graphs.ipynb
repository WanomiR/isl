### Setup

```python
import numpy as np
from matplotlib import pyplot as plt

rng = np.random.default_rng(2_147_483_647)

plt.style.use("dark_background")
plt.rcParams["figure.figsize"] = 6, 4
```

### Plot / Scatter plot

```python
fig, ax = plt.subplots()

x = rng.standard_normal(100)
y = rng.standard_normal(100)

ax.plot(x, y, "o");
```

### Subplots

```python
fig, axs = plt.subplots(2, 3, figsize=(8, 4))

axs[0, 1].plot(x, y, "o")
axs[0, 1].set_xlim([-1, 1])

axs[1, 2].plot(x, y, marker="+")

fig.set_size_inches(8, 4)
fig.tight_layout()
```

### Contour

The `ax.contour()` method produces a _contour plot_ in order to represent three-dimensional data, similar to topographical map.

```python
fig, ax = plt.subplots()

x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.multiply.outer(np.cos(y), 1 / (1 + x**2))

ax.contour(x, y, z);
```

```python
fig, ax = plt.subplots()

ax.contour(x, y, z, levels=30); # add more levels to increase resolution
```

### Heatmap

```python
fig, ax = plt.subplots()

ax.imshow(z);
```
